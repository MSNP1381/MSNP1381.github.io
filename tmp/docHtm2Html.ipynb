```python
import re
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
from bs4 import Tag
import json
import os
import cssutils
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
```

```python
with open("nistam Bolbol.htm") as fp:
    soup = BeautifulSoup(fp, "html.parser")
```

```python
l=soup.select('span')
```

```python
def get_download_link(link:str):
    chrome_driver_path = 'c:\\chromedriver.exe'
    service = Service(executable_path=chrome_driver_path)
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--no-sandbox')
    driver = webdriver.Chrome(service=service,options=chrome_options)  
    driver.get(link)

    # Wait for dynamic content to load
    wait = WebDriverWait(driver, 3)
    # element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.jjoUKC:nth-child(1) > button:nth-child(1)')))

    # Extract HTML source code
    html = driver.page_source

    new_soup = BeautifulSoup(html, 'html.parser')
    os.makedirs("pages_html",exist_ok=True)
    with open(os.path.join("pages_html",link.split('/')[-1]),'w',encoding='utf-8') as f:
        f.write(new_soup.prettify())
    driver.quit()
      
```

```python
time_format=r"^(?:[0-5][0-9]):[0-5][0-9]$"
for i in l:

    if re.match( time_format,i.text):
        whole_time=i.parent
        text_el=i.parent.next_sibling.next
        tag=soup.new_tag("a",href=f'#{i.text}' )
        tag.attrs['style']="text-decoration: none;font-family:Vazir"
        text_el.attrs['id']=i.text
        sheet=cssutils.parseStyle(text_el.attrs['style'])
        # sheet.fontFamily= "Calibri,inherit"
        # sheet.color= "inherit"
        text_el['style'] = sheet.cssText
        text_el.wrap(tag)
        whole_time.extract()
```

```python
with open("parsed_html.html",'w',encoding="utf-8") as f:
    f.write(soup.prettify())
```
